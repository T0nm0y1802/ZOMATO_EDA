```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns
```

```python
df_country.info()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
 country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

## Observation
 1.rating 4.5 to 4.9---> Excellent
2.rating 4.0 to 4.4---> very good
3.rating 3.9 to 3.5---> good
4.rating 3.4 to 2.5---> average
5.rating 2.4 and below---> poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating text',palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(data=ratings,x='Rating color',palette=['blue','red','orange','yellow','green','green'])
```
